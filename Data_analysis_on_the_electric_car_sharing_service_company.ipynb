```
#Importing the pandas library
#
import pandas as pd
```

```
#Importing the numpy library
#
import numpy as np
```

```
# Making the dataframe from the csv file
#
df = pd.read_csv('/content/Autolib_dataset (2).csv')
df 
```

```
#Finding the number  of missing values in each column
#
df.isnull().sum()
```

```
#Finding duplicated values in the datset
#
df.duplicated().sum()
```

```
#Changing the values in the kind column to lowercase
#
df['Kind'] = df['Kind'].str.lower()
df
```

```
#Identifying the most popular hour of the day for picking up a shared electric car(Bluecar) in the city of paris
# over the month of april 2018
#
df1 = df.loc[(df['year'] == '2018')& (df['City'] == 'Paris')& (df['month'] == '4')]
diff_values = df1['Bluecar counter'].diff(periods = 1)
df1['diff_values'] = diff_values
df1.groupby(['diff_values']).min().head(1)
```

```
#What is the most popular hour for returning cars
#
df2 = df['Cars'].diff(periods = 1)
df['df2'] = df2
df.groupby('df2').max().head(1)
```

```
# What station is the most popular
# Overall
#
df.groupby(['ID'])['Station type'].count().sort_values(ascending = False).head(1)
```

```
# What station is the most popular
# At the most popular picking hour
#
diff_values = df['Cars'].diff(periods = 1)
df['diff_values'] = diff_values
df.groupby(['diff_values']).min().head(1)
```

```
# What postal code is the most popular for picking up bluecars? Does the most popular station belong to that code?
# Overall
#
diff_values = df['Bluecar counter'].diff(periods = 1)
df['diff_values'] = diff_values
df.groupby(['diff_values']).min().head(1)
# The most popular station does not belong to that code
```

```
# What postal code is the most popular for picking up bluecars? Does the most popular station belong to that code?
# At the most popular picking hour
#
diff_values = df['Bluecar counter'].diff(periods = 1)
df['diff_values'] = diff_values
df.groupby(['diff_values']).min().head(1)
# The most popular postalcode does not belong to the most popular station
```

```
#Do the results change if you consider Utilib and Utilib 1.4 instead of Bluecars? 
#
diff_values = df['Utilib counter'].diff(periods = 1)
df['diff_values'] = diff_values
df.groupby(['diff_values']).min().head(1)
# The postal code is 75004 and the station is bagneux-verdun-127
#

diff_values = df['Utilib 1.4 counter'].diff(periods = 1)
df['diff_values'] = diff_values
df.groupby(['diff_values']).min().head(1)
#The postal code is 75002 and the Station is paris-paix-24
#
```
